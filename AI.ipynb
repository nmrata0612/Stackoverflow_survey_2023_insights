```python
import pandas as pd
import numpy as np
```

```python
AIDf = df[['AISearchHaveWorkedWith', 'AISearchWantToWorkWith', 'ResponseId']]
```

```python
AIdf = required_data.merge(AIdf, how='left', on='ResponseId')
```

```python
AIDf = AIDf.dropna()
```

```python
def process_AI(row):
    list1 = row['AISearchHaveWorkedWith']
    list2 = row['AISearchWantToWorkWith']
    list3 = set(list1) & set(list2)
    return list(list3)

AIDf['continueAI'] = AIDf.apply(process_AI, axis=1)
```

```python
def process_AI(row):
    list1 = row['AISearchHaveWorkedWith']
    list2 = row['AISearchWantToWorkWith']
    list3 = set(list2) - set(list1)
    return list(list3)

AIDf['newAI'] = AIDf.apply(process_AI, axis=1)
```

```python
AIDf.head()
```

```python
AIuse = AIDf[['continueAI', 'ResponseId']]
AIuse = AIuse.explode('continueAI')
```

```python
AIuse = AIuse.groupby(by='continueAI', as_index = False)['ResponseId'].count()
```

```python
AIuse
```

```python
desireAI = AIDf[['newAI', 'ResponseId']]
desireAI = desireAI.explode('newAI')
```

```python
desireAI = desireAI.groupby(by='newAI', as_index=False)['ResponseId'].count()
```

```python
AITable = AIuse.merge(desireAI, how='inner', left_on='continueAI', right_on='newAI')
```

```python
AITable = AITable[['continueAI', 'ResponseId_x', 'ResponseId_y']]
AITable.rename(columns={
    'continueAI':'AI',
    'ResponseId_x':'CurrentlyUsed',
    'ResponseId_y':'DesiredToUse'
}, inplace = True)
AITable
```

```python
AITable.to_csv('AITable.csv')
```
