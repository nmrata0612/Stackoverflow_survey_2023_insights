```python
DevAIdf = df[['AIDevWantToWorkWith','AIDevHaveWorkedWith', 'ResponseId']]
```

```python
DevAIdf = required_data.merge(DevAIdf, how='left', on='ResponseId')
```

```python
DevAIdf = DevAIdf.dropna()
```

```python
def process_DevAI(row):
    list1 = row['AIDevHaveWorkedWith']
    list2 = row['AIDevWantToWorkWith']
    list3 = set(list1) & set(list2)
    return list(list3)

DevAIdf['continueDevAI'] = DevAIdf.apply(process_DevAI, axis=1)
```

```python
def process_DevAI(row):
    list1 = row['AIDevHaveWorkedWith']
    list2 = row['AIDevWantToWorkWith']
    list3 = set(list2) - set(list1)
    return list(list3)

DevAIdf['newDevAI'] = DevAIdf.apply(process_DevAI, axis=1)
```

```python
DevAIdf.head()
```

```python
DevAIuse = DevAIdf[['continueDevAI', 'ResponseId']]
DevAIuse = DevAIuse.explode('continueDevAI')
```

```python
DevAIuse = DevAIuse.groupby(by='continueDevAI', as_index=False)['ResponseId'].count()
```

```python
desireDevAI = DevAIdf[['newDevAI', 'ResponseId']]
desireDevAI = desireDevAI.explode('newDevAI')
```

```python
desireDevAI = desireDevAI.groupby(by='newDevAI', as_index=False)['ResponseId'].count()
```

```python
DevAITable = DevAIuse.merge(desireDevAI, how='inner', left_on='continueDevAI', right_on='newDevAI')
```

```python
DevAITable = DevAITable[['continueDevAI', 'ResponseId_x', 'ResponseId_y']]
DevAITable.rename(columns={
    'continueDevAI':'DevAI',
    'ResponseId_x':'CurrentlyUsed',
    'ResponseId_y':'DesiredToUse'
}, inplace=True)
DevAITable
```

```python
DevAITable.to_csv('DevAITable.csv')
```
