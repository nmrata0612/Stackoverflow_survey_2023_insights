```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('survey_results_public.csv')
```

```python
required_data = pd.read_csv("required_data.csv")
```

```python
df['LanguageHaveWorkedWith'] = df['LanguageHaveWorkedWith'].str.split(';')
df['LanguageWantToWorkWith'] = df['LanguageWantToWorkWith'].str.split(';')
```

```python
LanguagesDF = required_data.merge(LanguagesDF, how='left', on='ResponseId')
```

```python
LangHaveWorkedDF = LanguagesDF[['ResponseId', 'LanguageHaveWorkedWith']]
```

```python
LangHaveWorkedDF = LangHaveWorkedDF.explode('LanguageHaveWorkedWith')
```

```python
LangHaveWorkedDF['LanguageHaveWorkedWith'] = LangHaveWorkedDF['LanguageHaveWorkedWith'].str.replace(r"\[|\]", "", regex=True)
```
